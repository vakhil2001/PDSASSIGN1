```python
#import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Cleaned_StudentsPerformance.csv') #read the csv data
```

```python
df.head()
```

# Five Different visualizations.



#**1.BarPlot**

```python
# Bar plot for gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# **2.Scatter Plot of Math vs. Reading Scores**

```python
# Scatter plot for math vs reading scores
plt.figure(figsize=(8,6))
sns.scatterplot(x='math score', y='reading score', hue='gender', data=df, palette='coolwarm', s=100)
plt.title('Math vs Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

# **3. Histogram of Parental Level of Education**

```python
# Histogram for parental level of education
plt.figure(figsize=(8,6))
sns.histplot(df['parental level of education'], bins=6, color='green', kde=True)
plt.title('Distribution of Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

# **4.Point Plot of Test Scores by Lunch Type**

```python
# Create a new column for total score
df['total score'] = df['math score'] + df['reading score'] + df['writing score']

# Violin plot for total scores by lunch type
plt.figure(figsize=(8,6))
sns.violinplot(x='lunch', y='total score', data=df, palette='muted')
plt.title('Total Scores by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Total Score')
plt.show()
```

# **5.Box Plot of Math Scores by Gender**

```python
# Box plot for math scores by gender
plt.figure(figsize=(8,6))
sns.boxplot(x='gender', y='math score', data=df, palette='Set3')
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```
