```python
import pandas as pd  #importing the required libraires
```

# Load the raw dataset

```python
df = pd.read_csv('/content/sample_data/datasetRaw.csv')
df
```

# Then Convert the 'Frailty' column from 'N'/'Y' to 0/1 or boolean value

```python
df.sort_values(by='Age', inplace=True) #sort the age coloumn

df['Frailty'].replace(['Y', 'N'], [0, 1], inplace=True)
```

```python
df
```

# let's Check for missing values in the dataset

```python
missing_values = df.isnull().sum()
```

```python
print("Missing values in the dataset:\n", missing_values)
```

# Save the cleaned dataset to a new CSV file

```python
df.to_csv('CleanedDataSet.csv', index=False)

print("Then file will be saved as 'CleanedDataSet.csv'.")
```
