```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv('Frailty_Cleaned_Dataset.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Get summary statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Boxplot for grip strength by frailty status

```python
sns.boxplot(x='Frailty', y='Grip strength', data=df)
plt.title('Grip Strength by Frailty')
plt.xlabel('Frailty (0 = No, 1 = Yes)')
plt.ylabel('Grip Strength (kg)')
plt.show()
```

```python
import matplotlib.pyplot as plt
```

# Create the pie chart

```python
frailty_counts = df['Frailty'].value_counts()
labels = ['Non-Frail', 'Frail']
plt.pie(frailty_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.axis('equal')
plt.title('Distribution of Frailty')
plt.show()
```

# Create a line graph for Age vs Grip Strength

```python
df_sorted = df.sort_values('Age')
plt.plot(df_sorted['Age'], df_sorted['Grip strength'], marker='o', linestyle='-', color='b')
plt.title('Grip Strength vs Age')
plt.xlabel('Age (years)')
plt.ylabel('Grip Strength (kg)')
plt.show()
```

# Calculate the correlation matrix

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```
